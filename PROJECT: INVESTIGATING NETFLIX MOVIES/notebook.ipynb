# **Base code**

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv", index_col = 0)
```

```python
netflix_df_90s = netflix_df.copy()
```

```python
netflix_df_90s = netflix_df_90s[(netflix_df_90s['release_year'] >= 1990) & (netflix_df_90s['release_year'] <= 1999)]
```

# Exploratory Analysis

### The mean of duration

```python
mean_of_duration = netflix_df_90s['duration'].mean()
print(mean_of_duration)
```

### The mode of duration

```python
mode_of_duration = netflix_df_90s['duration'].mode()[0]
print(mode_of_duration)
```

### What was the most frequent movie duration in the 1990s?

```python
duration = mode_of_duration
print(f" The most frequente movie duration was : {duration}")
```

### Actions Movies

### Conte o número de filmes de ação curtos lançados na década de 1990 e salve esse número inteiro como short_movie_count.

```python
action_movies = netflix_df_90s[netflix_df_90s['genre'] == 'Action']
```

```python
action_movies.head()
```

```python
short_movie_count = action_movies[action_movies['duration'] < 90].shape[0]
```

```python
print(short_movie_count)
```
